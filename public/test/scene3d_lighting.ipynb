```python
import scene3d, math

scene = scene3d.Scene()
scene.set_sky('#0a0a1a')  # deep night sky

# Dim the ambient light so point lights really pop
scene.ambient.set_brightness(8)
scene.ambient.set_color('#1a1a4a')  # cool blue tint for the little ambient that remains

# Ground
ground = scene.set_ground(length=30, width=30)
ground.set_material(scene3d.Material.Marble.Black)
ground.set_tiling(6)

# A central sphere to catch the coloured light
orb = scene3d.Shapes.Sphere(diameter=1.5, segments=32)
orb.set_position(0, 0.75, 0)
orb.set_material(scene3d.Material.Marble.Gray)
orb.set_glossiness(0.8)
scene.add(orb)

# A ring of smaller shapes around the centre
ring_shapes = []
colors = ['#cc2200', '#22cc00', '#0022cc', '#cc8800', '#cc00cc']
for i in range(5):
    angle = i * (360 / 5)
    rad = math.radians(angle)
    x = math.cos(rad) * 5
    z = math.sin(rad) * 5
    shape = scene3d.Shapes.Box(width=0.8, height=1.5, depth=0.8)
    shape.set_position(x, 0.75, z)
    shape.set_color(colors[i])
    scene.add(shape)
    ring_shapes.append(shape)

# Three coloured point lights that orbit the scene
red_light   = scene.add_light(6, 3, 0)
green_light = scene.add_light(-3, 3, 5.2)
blue_light  = scene.add_light(-3, 3, -5.2)

red_light.set_color('#ff2200').set_brightness(100)
green_light.set_color('#00ff44').set_brightness(100)
blue_light.set_color('#2244ff').set_brightness(100)

# Show the indicator spheres so you can see the lights orbiting
red_light.set_visible(True)
green_light.set_visible(True)
blue_light.set_visible(True)

ORBIT_RADIUS = 6
ORBIT_HEIGHT = 3
ORBIT_SPEED  = 0.4  # radians per second

ctx = scene.get_context('2d')
t = 0.0

@scene.on_frame
def animate(dt):
    global t
    t += dt

    # Orbit the three lights 120 degrees apart
    for i, light in enumerate([red_light, green_light, blue_light]):
        a = t * ORBIT_SPEED + i * (2 * math.pi / 3)
        light.set_position(
            math.cos(a) * ORBIT_RADIUS,
            ORBIT_HEIGHT,
            math.sin(a) * ORBIT_RADIUS,
        )

    # Slowly spin the central orb
    orb.set_rotation(y=t * 20)

    ctx.clear()
    ctx.fill_style = 'rgba(255,255,255,0.7)'
    ctx.font = '14px sans-serif'
    ctx.fill_text(f't = {t:.1f}s  |  orbit speed: {ORBIT_SPEED:.1f} rad/s', 10, 24)

scene.camera.set_distance(18)
scene.camera.look_at(0, 1, 0)

scene.run()
```
