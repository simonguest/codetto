# Collision Detection & Key Events

Use the arrow keys to move the red ball. When it hits a coloured box, the box turns orange and the hit counter goes up.

Demonstrates `on_key` (arrow key movement) and `on_collide` (two separate collision handlers on the same mesh).

```python
import scene3d

scene = scene3d.Scene()
scene.set_sky("#87CEEB")

ground = scene.set_ground(length=24, width=24)
ground.set_material(scene3d.Material.Grass.Bright)
ground.set_tiling(6)

# Blue target box
blue_box = scene3d.Shapes.Box(width=2, height=2, depth=2)
blue_box.set_position(-4, 1, 0)
blue_box.set_color("#4488ff")
scene.add(blue_box)

# Green target box
green_box = scene3d.Shapes.Box(width=2, height=2, depth=2)
green_box.set_position(4, 1, 0)
green_box.set_color("#44bb44")
scene.add(green_box)

# Player ball — moved with arrow keys
ball = scene3d.Shapes.Sphere(diameter=1)
ball.set_position(0, 0.5, 0)
ball.set_color("#ff4444")
scene.add(ball)

hits = [0]

ball.on_collide(blue_box,  lambda: (blue_box.set_color("#ff8800"),  hits.__setitem__(0, hits[0] + 1)))
ball.on_collide(green_box, lambda: (green_box.set_color("#ff8800"), hits.__setitem__(0, hits[0] + 1)))

speed = 5.0
move = {'x': 0, 'z': 0}

scene.on_key(scene3d.Key.LEFT,  lambda: move.update({'x': -speed}))
scene.on_key(scene3d.Key.RIGHT, lambda: move.update({'x':  speed}))
scene.on_key(scene3d.Key.UP,    lambda: move.update({'z':  speed}))
scene.on_key(scene3d.Key.DOWN,  lambda: move.update({'z': -speed}))

ctx = scene.get_context('2d')

@scene.on_frame
def animate(dt):
    x, y, z = ball.get_position()
    x = max(-11, min(11, x + move['x'] * dt))
    z = max(-11, min(11, z + move['z'] * dt))
    ball.set_position(x, y, z)
    move['x'] = 0
    move['z'] = 0
    ctx.clear()
    ctx.fill_style = '#ffffff'
    ctx.font = 'bold 22px sans-serif'
    ctx.fill_text(f'Hits: {hits[0]}  |  Arrow keys to move', 12, 32)

scene.run()
```
