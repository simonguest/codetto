# Collision Detection

A red ball rolls back and forth across the scene. When it hits a coloured box, the box turns orange and the hit counter goes up.

This demonstrates `on_collide` — two separate collision handlers are registered on the same ball mesh.

```python
import scene3d

scene = scene3d.Scene()
scene.set_sky("#87CEEB")

ground = scene.set_ground(length=24, width=24)
ground.set_material(scene3d.Material.Grass.Bright)
ground.set_tiling(6)

# Blue target box
blue_box = scene3d.Shapes.Box(width=2, height=2, depth=2)
blue_box.set_position(-3, 1, 0)
blue_box.set_color("#4488ff")
scene.add(blue_box)

# Green target box
green_box = scene3d.Shapes.Box(width=2, height=2, depth=2)
green_box.set_position(3, 1, 0)
green_box.set_color("#44bb44")
scene.add(green_box)

# Moving ball
ball = scene3d.Shapes.Sphere(diameter=1)
ball.set_position(-9, 0.5, 0)
ball.set_color("#ff4444")
scene.add(ball)

hits = [0]

def on_hit_blue():
    hits[0] += 1
    blue_box.set_color("#ff8800")

def on_hit_green():
    hits[0] += 1
    green_box.set_color("#ff8800")

ball.on_collide(blue_box, on_hit_blue)
ball.on_collide(green_box, on_hit_green)

ctx = scene.get_context('2d')
speed = 6.0
direction = [1]

@scene.on_frame
def animate(dt):
    x, y, z = ball.get_position()
    x += speed * direction[0] * dt
    if x > 9:
        direction[0] = -1
    elif x < -9:
        direction[0] = 1
    ball.set_position(x, y, z)
    ctx.clear()
    ctx.fill_style = '#ffffff'
    ctx.font = 'bold 22px sans-serif'
    ctx.fill_text(f'Hits: {hits[0]}', 12, 32)

scene.run()
```
