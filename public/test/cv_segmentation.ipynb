```python
import graphics
import cv
import time

canvas = graphics.canvas()
camera = cv.start_camera(canvas=canvas)
segmenter = cv.start_segmenter(camera)

# Paint hair red at 50% opacity
cv.color_segment(canvas, segmenter, cv.SEGMENT.HAIR, '#ff0000', 0.5)

try:
  while True:
    segments = segmenter.get_segments()
    print(f'Segments: {segments}')
    time.sleep(0.5)
finally:
  segmenter.stop()
  camera.stop()
```
