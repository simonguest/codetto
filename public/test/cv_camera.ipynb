```python
import cv
import time

# Headless camera - no canvas required
camera = cv.start_camera()
time.sleep(0.5)  # let rAF loop run a few ticks
camera.stop()
print('Headless camera OK')
```

```python
import graphics
import cv
import time

# Camera with canvas - frames rendered to the output cell
canvas = graphics.canvas(640, 480)
camera = cv.start_camera(canvas=canvas)
time.sleep(0.5)  # let rAF loop draw a few frames on the main thread
camera.stop()
print('Camera with canvas OK')
```
