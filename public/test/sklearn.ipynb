# sklearn Test

Tests that `scikit-learn` works and that its matplotlib output renders as a PNG image.

```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_checkerboard

data, rows, columns = make_checkerboard(
    shape=(30, 30), n_clusters=(4, 3), noise=10, shuffle=False, random_state=42
)

plt.matshow(data, cmap=plt.cm.Blues)
plt.title("Checkerboard")
plt.show()
```
