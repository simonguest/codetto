```python
import graphics
import cv
import time

canvas = graphics.canvas()
camera = cv.start_camera(canvas=canvas)
detector = cv.start_gesture_detector(camera)

try:
  while True:
    hands = detector.get_detections()
    canvas.draw_hands(hands)
    print(f'Hands detected: {len(hands)}')
    for hand in hands:
      tip = hand['landmarks'][cv.HAND.INDEX_FINGER_TIP]
      print(f"{hand['handedness']} hand: {hand['gesture']} ({hand['confidence']}) tip at ({tip['x']}, {tip['y']})")
    time.sleep(0.5)
finally:
  detector.stop()
  camera.stop()
```
