```python
import scene3d

scene = scene3d.Scene()

box = scene3d.Shapes.Box()
box.set_position(0, 0.5, 0)
scene.add(box)

scene.on_key(scene3d.Key.LEFT,  lambda: print('key: left'))
scene.on_key(scene3d.Key.RIGHT, lambda: print('key: right'))
scene.on_key(scene3d.Key.UP,    lambda: print('key: up'))
scene.on_key(scene3d.Key.DOWN,  lambda: print('key: down'))
scene.on_key(scene3d.Key.SPACE, lambda: print('key: space'))
scene.on_key('w',               lambda: print('key: w'))

print('starting')  # ensures the stdout tab is rendered before scene.run() blocks
scene.run()
```
