```python
import graphics
import cv
import time

canvas = graphics.canvas()
camera = cv.start_camera(canvas=canvas)
detector = cv.start_face_detector(camera)

try:
    while True:
        faces = detector.get_detections()
        canvas.draw_bounding_boxes(faces)
        print(f'Faces detected: {len(faces)}')
        time.sleep(0.5)
finally:
    detector.stop()
    camera.stop()
```
