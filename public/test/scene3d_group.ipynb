```python
import scene3d

scene = scene3d.Scene()
scene.set_sky(scene3d.Sky.DEEP_SPACE)

# Build a satellite from grouped meshes.
# All parts rotate together as one unit to verify Group is working.
satellite = scene3d.Group()

body = scene3d.Shapes.Box(width=0.8, height=0.8, depth=0.8)
body.set_material(scene3d.Material.Chip.CircuitBlue)

panel_l = scene3d.Shapes.Box(width=1.2, height=0.05, depth=0.6)
panel_l.set_position(-1.1, 0, 0)
panel_l.set_color('#1a3a6e')

panel_r = scene3d.Shapes.Box(width=1.2, height=0.05, depth=0.6)
panel_r.set_position(1.1, 0, 0)
panel_r.set_color('#1a3a6e')

antenna = scene3d.Shapes.Cylinder(diameter=0.05, height=0.8, tessellation=8)
antenna.set_position(0, 0.65, 0)
antenna.set_color('#aaaaaa')

satellite.add(body)
satellite.add(panel_l)
satellite.add(panel_r)
satellite.add(antenna)

scene.add(satellite)

t = 0.0

@scene.on_frame
def animate(dt):
    global t
    t += dt
    satellite.set_rotation(x=t * 10, y=t * 30)

scene.run()
```
