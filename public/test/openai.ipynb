# OpenAI Library Test

Check that the library is present and loading.

```python
import openai
import httpx
import os

OPENROUTER_API_KEY = os.environ["OPENROUTER_API_KEY"]

# Strip x-stainless-* headers — OpenRouter's CORS policy rejects them
def strip_stainless(request: httpx.Request) -> None:
    for k in [k for k in request.headers if k.startswith("x-stainless")]:
        del request.headers[k]

client = openai.OpenAI(
    base_url='https://openrouter.ai/api/v1',
    api_key=OPENROUTER_API_KEY,
    http_client=httpx.Client(event_hooks={'request': [strip_stainless]})
)

MODEL = 'openai/gpt-4.1-mini' #@param ["openai/gpt-4.1-mini", "anthropic/claude-sonnet-4.5", "google/gemini-2.5-pro"]

response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": "You help students with their geography lessons."},
        {"role": "user", "content": "Give me all of the US states that begin with the letter A."},
    ],
)

print(response.choices[0].message.content)
```
