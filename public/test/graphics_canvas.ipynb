```python
import graphics

canvas = graphics.canvas(400, 300)
ctx = canvas.get_context('2d')

# fill_rect and fill_style
ctx.fill_style = '#e74c3c'
ctx.fill_rect(10, 10, 80, 60)

# stroke_rect, stroke_style, line_width
ctx.stroke_style = '#2980b9'
ctx.line_width = 3
ctx.stroke_rect(110, 10, 80, 60)

# arc + fill (filled circle)
ctx.fill_style = '#27ae60'
ctx.begin_path()
ctx.arc(260, 40, 30, 0, 6.2832)
ctx.fill()

# arc + stroke (circle outline)
ctx.stroke_style = '#8e44ad'
ctx.line_width = 2
ctx.begin_path()
ctx.arc(360, 40, 30, 0, 6.2832)
ctx.stroke()

# fill_text and font
ctx.fill_style = '#2c3e50'
ctx.font = '18px sans-serif'
ctx.fill_text('Hello graphics!', 10, 120)

# clear_rect (punch a hole in the red rectangle)
ctx.clear_rect(30, 25, 40, 30)

print('Canvas drawing OK')
```
