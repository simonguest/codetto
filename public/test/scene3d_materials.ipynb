```python
import scene3d

scene = scene3d.Scene()

# Sky: HDR environment (also sets scene.environmentTexture for PBR IBL)
scene.set_sky(scene3d.Sky.PURE_SKY)
print('ok: sky env')

# Ground: returns a Mesh; apply material and tiling
ground = scene.set_ground(length=20, width=20)
ground.set_material(scene3d.Material.Grass.Bright)
ground.set_tiling(8)
print('ok: ground material + tiling')

# Box: PBR material + glossiness set BEFORE scene.add()
box = scene3d.Shapes.Box(width=1, height=1, depth=1)
box.set_position(0, 0.5, 0)
box.set_material(scene3d.Material.Bricks.DarkClay)
box.set_glossiness(0.4)
scene.add(box)
print('ok: box pbr pre-add')

# Sphere: simple texture (planets) set AFTER scene.add()
sphere = scene3d.Shapes.Sphere(diameter=1, segments=16)
sphere.set_position(2.5, 0.5, 0)
scene.add(sphere)
sphere.set_material(scene3d.Material.Planets.Earth)
print('ok: sphere simple material post-add')

# Cylinder: marble + high glossiness + non-square tiling
cylinder = scene3d.Shapes.Cylinder(diameter=1, height=2, tessellation=24)
cylinder.set_position(-2.5, 1, 0)
cylinder.set_material(scene3d.Material.Marble.Gray)
cylinder.set_glossiness(0.8)
cylinder.set_tiling(3, 2)
scene.add(cylinder)
print('ok: cylinder material + glossiness + tiling')

# Replace material on an already-added mesh
box.set_material(scene3d.Material.Wood.Oak)
print('ok: material replacement post-add')

# Tiling set before material replacement — should survive the new material
sphere.set_tiling(4)
sphere.set_material(scene3d.Material.Marble.Black)
print('ok: tiling persists across material replacement')

# Sky switch: flat colour clears environmentTexture
scene.set_sky('#1a1a2e')
print('ok: sky switched to flat colour')

print('scene3d materials: all ok')
# scene.run()
```
