```python
import scene3d
import math

scene = scene3d.Scene()
scene.set_sky("#87CEEB")
scene.set_ground(length=10, width=10)

box = scene3d.Shapes.Box(width=1, height=1, depth=1)
box.set_position(0, 0.5, 0)
box.set_color("#cc4400")

sphere = scene3d.Shapes.Sphere(diameter=1, segments=16)
sphere.set_position(2, 0.5, 0)
sphere.set_color("#0044cc")

last_clicked = "nothing"

def on_box_click():
    global last_clicked
    box.set_color("#ff6600")
    last_clicked = "box"

def on_sphere_click():
    global last_clicked
    sphere.set_color("#00aaff")
    last_clicked = "sphere"

box.on_click(on_box_click)
sphere.on_click(on_sphere_click)

scene.add(box)
scene.add(sphere)

ctx = scene.get_context('2d')
t = 0.0

@scene.on_frame
def animate(dt):
    global t
    t += dt
    box.set_position(0, 0.5 + math.sin(t * 2) * 1.0, 0)

    ctx.clear()
    ctx.fill_style = '#ffffff'
    ctx.font = 'bold 16px sans-serif'
    ctx.fill_text(f'Time: {t:.1f}s  |  Last clicked: {last_clicked}', 10, 24)

print("scene3d: scene created")
# scene.run()
```
