```python
import scene3d
import math

scene = scene3d.Scene()
scene.set_sky("#87CEEB")
scene.set_ground(length=10, width=10)

box = scene3d.Shapes.Box(width=1, height=1, depth=1)
box.set_position(0, 0.5, 0)
box.set_color("#cc4400")

sphere = scene3d.Shapes.Sphere(diameter=1, segments=16)
sphere.set_position(2, 0.5, 0)
sphere.set_color("#0044cc")

plane = scene3d.Shapes.Plane(width=1.5, height=1.5)
plane.set_position(-2, 0.75, 0)
plane.set_rotation(y=30)
plane.set_color("#44aa00")

cone = scene3d.Shapes.Cone(diameter=1, height=1.5, tessellation=24)
cone.set_position(-4, 0.75, 0)
cone.set_color("#aa00aa")

torus = scene3d.Shapes.Torus(diameter=1, thickness=0.35, tessellation=24)
torus.set_position(4, 0.5, 0)
torus.set_color("#cc9900")

label = scene3d.Shapes.Text("Hello!", size=0.8, depth=0.15)
label.set_position(0, 2.5, 0)
label.set_color("#ffffff")

last_clicked = "nothing"

def on_box_click():
    global last_clicked
    box.set_color("#ff6600")
    last_clicked = "box"

def on_sphere_click():
    global last_clicked
    sphere.set_color("#00aaff")
    last_clicked = "sphere"

box.on_click(on_box_click)
sphere.on_click(on_sphere_click)

scene.add(box)
scene.add(sphere)
scene.add(plane)
scene.add(cone)
scene.add(torus)
scene.add(label)

ctx = scene.get_context('2d')
t = 0.0

@scene.on_frame
def animate(dt):
    global t
    t += dt
    box.set_position(0, 0.5 + math.sin(t * 2) * 1.0, 0)
    torus.set_rotation(y=t * 60)

    ctx.clear()
    ctx.fill_style = '#ffffff'
    ctx.font = 'bold 16px sans-serif'
    ctx.fill_text(f'Time: {t:.1f}s  |  Last clicked: {last_clicked}', 10, 24)

print("scene3d: scene created")
# scene.run()
```
