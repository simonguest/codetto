```python
import graphics
import cv
import time

canvas = graphics.canvas()
camera = cv.start_camera(canvas=canvas)
detector = cv.start_pose_detector(camera)

try:
  while True:
    poses = detector.get_detections()
    canvas.draw_poses(poses)
    print(f'Poses detected: {len(poses)}')
    for pose in poses:
      left_wrist = pose[cv.POSE.LEFT_WRIST]
      right_wrist = pose[cv.POSE.RIGHT_WRIST]
      if left_wrist['visibility'] > 0.5:
        print(f'Left wrist: ({left_wrist["x"]}, {left_wrist["y"]})')
      if right_wrist['visibility'] > 0.5:
        print(f'Right wrist: ({right_wrist["x"]}, {right_wrist["y"]})')
    time.sleep(0.5)
finally:
  detector.stop()
  camera.stop()
```
