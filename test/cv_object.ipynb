```python
import graphics
import cv
import time

canvas = graphics.canvas()
camera = cv.start_camera(canvas=canvas)
detector = cv.start_object_detector(camera, delegate='CPU')

try:
    while True:
        objects = detector.get_detections()
        canvas.draw_bounding_boxes(objects)
        for obj in objects:
            print(f"{obj['type']}: {obj['confidence']}")
        time.sleep(0.5)
finally:
    detector.stop()
    camera.stop()
```
