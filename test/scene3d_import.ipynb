```python
import scene3d

scene = scene3d.Scene()
scene.set_sky('#87CEEB')

# Plain dicts — equivalent to what Pydantic model_dump() produces
meshes = [
    # Box: color + position
    {'type': 'Box', 'width': 4, 'height': 2.5, 'depth': 4,
     'position': [0, 1.25, 0], 'color': '#cc8844'},
    # Sphere: material string resolved via Category.Name
    {'type': 'Sphere', 'diameter': 1.5,
     'position': [3, 0.75, 0], 'material': 'Marble.Gray'},
    # Cylinder: rotation + color
    {'type': 'Cylinder', 'diameter': 0.5, 'height': 3,
     'position': [-3, 1.5, 0], 'rotation': [0, 45, 0], 'color': '#666666'},
    # Unknown type — must be silently skipped
    {'type': 'Cone', 'position': [0, 0, 0], 'color': '#ff0000'},
]

scene.import_meshes(meshes)
print('ok: dict meshes imported')

# Object with model_dump() — simulates a Pydantic model instance
class _FakeMesh:
    def model_dump(self):
        return {'type': 'Box', 'width': 1, 'height': 1, 'depth': 1,
                'position': [5, 0.5, 0], 'color': '#0044cc'}

scene.import_meshes([_FakeMesh()])
print('ok: model_dump() path works')

# Unknown material string — should not raise, just skip material
scene.import_meshes([{'type': 'Box', 'material': 'Nonexistent.Fake', 'color': '#aabbcc'}])
print('ok: bad material string skipped gracefully')

print('scene3d import_meshes: all ok')
# scene.run()
```
