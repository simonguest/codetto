```python
import scene3d

scene = scene3d.Scene()

box = scene3d.Shapes.Box(width=1, height=1, depth=1)

# Set properties before scene.add() and verify getters read back correctly
box.set_position(3, 1, -2)
box.set_rotation(x=30, y=45, z=0)
box.set_scale(2, 0.5, 1)
box.set_color('#ff0000')

assert box.get_position() == (3, 1, -2), f'got: {box.get_position()}'
print('ok: get_position before add')

assert box.get_rotation() == (30, 45, 0), f'got: {box.get_rotation()}'
print('ok: get_rotation before add')

assert box.get_scale() == (2, 0.5, 1), f'got: {box.get_scale()}'
print('ok: get_scale before add')

assert box.get_color() == '#ff0000', f'got: {box.get_color()}'
print('ok: get_color before add')

scene.add(box)

# Values should be unchanged immediately after scene.add()
assert box.get_position() == (3, 1, -2), f'got: {box.get_position()}'
print('ok: get_position after add')

assert box.get_rotation() == (30, 45, 0), f'got: {box.get_rotation()}'
print('ok: get_rotation after add')

# Set new values post-add and verify getters reflect the updates
box.set_position(0, 5, 10)
assert box.get_position() == (0, 5, 10), f'got: {box.get_position()}'
print('ok: get_position after set post-add')

box.set_rotation(y=90)
assert box.get_rotation() == (0, 90, 0), f'got: {box.get_rotation()}'
print('ok: get_rotation after set post-add')

box.set_scale(3, 3, 3)
assert box.get_scale() == (3, 3, 3), f'got: {box.get_scale()}'
print('ok: get_scale after set post-add')

box.set_color('#0000ff')
assert box.get_color() == '#0000ff', f'got: {box.get_color()}'
print('ok: get_color after set post-add')

print('scene3d getters: all ok')
# scene.run()
```
