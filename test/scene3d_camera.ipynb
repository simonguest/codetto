```python
import scene3d

scene = scene3d.Scene()

# Verify camera property exists
assert scene.camera is not None
print('ok: camera property exists')

# Add a mesh to use as look_at / follow target
target = scene3d.Shapes.Box(width=1, height=1, depth=1)
target.set_position(3, 1, 0)
scene.add(target)

# set_position
scene.camera.set_position(0, 10, -15)
print('ok: set_position')

# move
scene.camera.move(dx=2, dy=0, dz=0)
print('ok: move')

# look_at with world coordinates
scene.camera.look_at(0, 0, 0)
print('ok: look_at coordinates')

# look_at with a mesh
scene.camera.look_at(target)
print('ok: look_at mesh')

# set_distance
scene.camera.set_distance(20)
print('ok: set_distance')

# follow
scene.camera.follow(target, distance=15)
print('ok: follow')

# stop following
scene.camera.follow(None)
print('ok: follow None (stop)')

# reset
scene.camera.reset()
print('ok: reset')

# method chaining
scene.camera.set_position(5, 5, -10).set_distance(18).look_at(0, 1, 0)
print('ok: method chaining')

print('scene3d camera: all ok')
```
