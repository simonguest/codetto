```python
import audio

# Single note: middle C for half a second
audio.play_note('C4', 0.5)
print('Single note OK')
```

```python
import audio

# Sharps and flats
audio.play_note('F#4', 0.4)
audio.play_note('Bb4', 0.4)
print('Accidentals OK')
```

```python
import audio

# C major chord
audio.play_note(['C4', 'E4', 'G4'], 1.0)
print('Chord OK')
```

```python
import audio

# Twinkle Twinkle Little Star (first phrase)
melody = [
    ('C4', 0.4), ('C4', 0.4),
    ('G4', 0.4), ('G4', 0.4),
    ('A4', 0.4), ('A4', 0.4),
    ('G4', 0.8),
    ('F4', 0.4), ('F4', 0.4),
    ('E4', 0.4), ('E4', 0.4),
    ('D4', 0.4), ('D4', 0.4),
    ('C4', 0.8),
]

audio.play_notes(melody)
print('Melody OK')
```

```python
import audio

# Melody ending on a chord
sequence = [
    ('E4', 0.3),
    ('G4', 0.3),
    ('B4', 0.3),
    (['E4', 'G4', 'B4'], 1.0),
]

audio.play_notes(sequence)
print('Melody with chord OK')
```

```python
import audio

# Async: play_note_async and play_notes_async return immediately
await audio.play_note_async('A4', 0.5)
print('play_note_async returned immediately')

await audio.play_notes_async([('C5', 0.3), ('E5', 0.3), ('G5', 0.3)])
print('play_notes_async returned immediately')
```

```python
import audio

# Climb across three octaves
notes = [(note, 0.25) for note in ['C3', 'C4', 'C5', 'C6']]
audio.play_notes(notes)
print('Multi-octave OK')
```
