# pygame Test

Tests that `pygame` renders a static frame as a PNG image via the custom display override.

```python
import pygame

pygame.init()
screen = pygame.display.set_mode((200, 200))
pygame.display.set_caption("Test")

# Draw a simple scene: blue background with a white rectangle
screen.fill((30, 100, 200))
pygame.draw.rect(screen, (255, 255, 255), (50, 50, 100, 100))

# flip() serialises the surface and sends it back as a PNG
pygame.display.flip()
pygame.quit()
```
