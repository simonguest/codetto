```python
import graphics
import cv
import time

canvas = graphics.canvas()
camera = cv.start_camera(canvas=canvas)
detector = cv.start_pose_detector(camera)

# get_context returns the student drawing layer (on top of the video/detection layer)
ctx = canvas.get_context('2d')

try:
  while True:
    poses = detector.get_detections()

    # draw_poses writes to the video layer and updates each camera frame
    canvas.draw_poses(poses)

    # clear only the student layer, then draw fresh text on top
    canvas.clear()
    ctx.fill_style = '#ffffff'
    ctx.font = '20px sans-serif'
    ctx.fill_text(f'Poses: {len(poses)}', 10, 30)

    for pose in poses:
      left_wrist = pose[cv.POSE.LEFT_WRIST]
      if left_wrist['visibility'] > 0.5:
        ctx.fill_style = '#00ffcc'
        ctx.fill_text(f'Wrist: ({left_wrist["x"]}, {left_wrist["y"]})', 10, 60)

    print(f'Poses: {len(poses)}')
    time.sleep(0.1)
finally:
  detector.stop()
  camera.stop()
```
